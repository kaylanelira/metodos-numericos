# Resolução da questão 6

> Estas equações são lineares e separáveis: Resolva (i) dy/dt = (y+4) cos(t) e
(ii) dy/dt = y exp(t)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from scipy.integrate import odeint
```

# Resolvendo (i) dy/dt = (y+4) cos(t)

## Solução Numérica

```python
# função dy/dt
def Q6i(y0, t):
  dydt = (y0 + 4) * np.cos(t);
  return dydt
```

Calculando a solução numérica

```python
# initial condition
y0 = 1 

#time points 
t = np.linspace(0, 20)  

# solve ODE
y_num = odeint(Q6i, y0, t)

#plot results 
plt.title('Q6(i) Numérica')
plt.plot(t, y_num) # ODE
plt.xlabel('tempo')
plt.ylabel('y(t)')
```

## Solução Analítica

```python
def Q6i_analitic(y0, t):
  dydt = ((y0 + 4) * np.exp(np.sin(t))) - 4 # sin(t0) = sin(0) = 0
  return dydt;
```

```python
#initial condition
y0 = 1

#time points 
t = np.linspace(0, 20) 

# expected function
y_ana  = Q6i_analitic(y0, t)

# plot results
plt.title('Q6(i) Analítica')
plt.plot(t, y_ana)
plt.xlabel('tempo')
plt.ylabel('y(t)')
```

## Comparando Analítica e Numérica

Comparando soluções numérica e analítica em um gráfico

```python
y_ana = Q6i_analitic(y0, t)
y_num = odeint(Q6i, y0, t)

# plot analitics results 
plt.plot(t,y_ana) 
plt.xlabel('time')
plt.ylabel('y(t)')

# plot numerical results 
plt.plot(t,y_num,'r:')
plt.legend(['y analítica','y numérica'])
plt.show()
```


# Resolvendo(ii) dy/dt = y exp(t)


## Solução Numérica

```python
def Q6ii(y0, t):
  dydt = y0 * np.exp(t);
  return dydt
```

Calculando a solução numérica

```python
# initial condition
y0 = 1 

#time points 
t = np.linspace(0, 2)  

# solve ODE
y_num = odeint(Q6ii, y0, t)

#plot results 
plt.title('Q6(ii) Numérica')
plt.plot(t, y_num) # ODE
plt.xlabel('tempo')
plt.ylabel('y(t)')
```

## Solução Analítica

```python
def Q6ii_analitic(y0, t):
  dydt = y0 * np.exp(np.exp(t) -  1)
  return dydt;
```

```python
#initial condition
y0 = 1

#time points 
t = np.linspace(0, 2) 

# expected function
y_ana  = Q6ii_analitic(y0, t)

# plot results
plt.title('Q6(ii) Analítica')
plt.plot(t, y_ana)
plt.xlabel('tempo')
plt.ylabel('y(t)')
```

## Comparando Analítica e Numérica

Comparando soluções numérica e analítica em um gráfico

```python
y_ana = Q6ii_analitic(y0, t)
y_num = odeint(Q6ii, y0, t)

# plot analitics results 
plt.plot(t,y_ana) 
plt.xlabel('time')
plt.ylabel('y(t)')

# plot numerical results 
plt.plot(t,y_num,'r:')
plt.legend(['y analítica','y numérica'])
plt.show()
```
